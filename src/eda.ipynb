### Import des librairies

```python
import pandas as pd
import plotly.express as px
```

```python
### Importation des librrairies
df_films = pd.read_csv("../data/films.csv")
df_films_fr = pd.read_csv("../data/films_fr.csv")
df_livres = pd.read_csv("../data/livres.csv")
df_musiques = pd.read_csv("../data/musiques.csv")
```

```python
# Concatenation des donn√©es
df_fusion  = pd.concat([df_films_fr, df_films, df_livres, df_musiques])
df_fusion.shape
```

```python
df_fusion.columns
```

```python
# Types de donn√©es
df_fusion.dtypes
```

```python
# separer les colonnes num√©riques des colonnes categorielles
num_cols = df_fusion.select_dtypes(exclude="object")
cat_cols = df_fusion.select_dtypes(include="object")
```

```python
# Verification des valeurs null
(df_fusion.isnull().sum() / len(df_fusion)) * 100
```

```python
# Gestion des valeurs manquantes
num_cols.drop(columns="annee", inplace=True, axis=1)
cat_cols.drop(columns="album", inplace=True, axis=1)
cat_cols.fillna("", inplace=True, axis=1)
```

```python
df = pd.concat([cat_cols, num_cols], axis=1)
df.isnull().sum()
```
